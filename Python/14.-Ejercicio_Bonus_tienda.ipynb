En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el monto final de una compra en una tienda en línea, después de aplicar un descuento. El programa debe hacer lo siguiente:

1. Solicita al usuario que ingrese el precio original de un artículo.

2. Pregunta al usuario si tiene un cupón de descuento (respuesta sí o no).

3. Si el usuario responde que sí, solicita que ingrese el valor del cupón de descuento.

4. Aplica el descuento al precio original del artículo, siempre y cuando el valor del cupón sea válido (es decir, mayor a cero). Por ejemplo, descuento de 15€. 

5. Muestra el precio final de la compra, teniendo en cuenta el descuento aplicado o sin él. 

6. Recuerda utilizar estructuras de control de flujo como if, elif y else para llevar a cabo estas acciones en tu programa de Python.

```python
print("Bienvenido a la tienda en línea.")

# Paso 1: Solicitar el precio original del artículo
precio_original = float(input("Por favor, introduce el precio original del artículo: €"))

# Paso 2: Preguntar si el usuario tiene un cupón de descuento
tiene_cupon = input("¿Tienes un cupón de descuento? (responde 'si' o 'no'): ").strip().lower()

# Paso 3 y 4: Verificar si hay cupón y calcular el descuento
if tiene_cupon == "si":
    valor_cupon = float(input("Introduce el valor del cupón de descuento (€): "))
    
    if valor_cupon > 0:
        # Calcular el precio final con el descuento aplicado
        precio_final = precio_original - valor_cupon
        
        # Asegurarnos de que el precio final no sea negativo
        if precio_final < 0:
            precio_final = 0  # El precio mínimo es 0
    else:
        print("El valor del cupón no es válido. Se aplicará el precio original.")
        precio_final = precio_original
else:
    # No hay cupón de descuento
    precio_final = precio_original

# Paso 5: Mostrar el precio final de la compra
print(f"El precio final de la compra es: {precio_final:.2f}€")
```

### Crea un programa nuevo que permita al usuario ingresar varios artículos, cada uno con su precio original y cupón de descuento. Aplica los descuentos correspondientes a cada producto y muestra el monto total final de la compra.

```python
print("Bienvenido a la tienda en línea.")
print("Podrás ingresar varios artículos con sus precios y descuentos.")
print("Cuando termines de ingresar los artículos, escribe 'fin'.")

# Inicializar variables
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del artículo
    articulo = input("Introduce el nombre del artículo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de artículos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del artículo
    try:
        precio_original = float(input(f"Introduce el precio original del artículo '{articulo}': €"))
    except ValueError:
        print("Error: Por favor, introduce un número válido para el precio.")
        continue

    # Paso 3: Preguntar si el artículo tiene un porcentaje de descuento
    tiene_descuento = input(f"¿El artículo '{articulo}' tiene un descuento? (responde 'si' o 'no'): ").strip().lower()

    # Inicializar precio final para este artículo
    precio_final = precio_original

    if tiene_descuento == "si":
        try:
            # Paso 4: Solicitar el porcentaje de descuento
            porcentaje_descuento = float(input("Introduce el porcentaje de descuento (por ejemplo, 30 para 30%): "))

            # Aplicar validación del porcentaje de descuento (mínimo 5%, máximo 50%)
            if porcentaje_descuento < 5:
                print("El descuento es menor al mínimo permitido (5%). Se aplicará el descuento mínimo del 5%.")
                porcentaje_descuento = 5
            elif porcentaje_descuento > 50:
                print("El descuento supera el máximo permitido (50%). Se aplicará el descuento máximo del 50%.")
                porcentaje_descuento = 50

            # Calcular el descuento y el precio final
            descuento = precio_original * (porcentaje_descuento / 100)
            precio_final -= descuento

        except ValueError:
            print("Error: Por favor, introduce un número válido para el porcentaje de descuento.")
            continue

    # Agregar el precio final del artículo al monto total
    monto_total += precio_final
    print(f"\nEl precio final del artículo '{articulo}' es: €{precio_final:.2f}\n")

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: €{monto_total:.2f}")
```

### Crea un nuevo programa que permita aplicar descuentos por categorías. En este caso, nuestra tienda aplicará descuentos en función de la categoría (por ejemplo, electrónica, ropa, juguetes). Pregunta al usuario a que categoría pertenece su producto y aplica el descuento correspondiente a la categoría. Debes solicitar al usuario que ingrese la categoría del articulo y el precio para aplicarle el descuento. 

```python
print("Bienvenido a la tienda en línea.")
print("Nuestra tienda ofrece descuentos según las categorías:")
print("1. Electrónica: 10% de descuento")
print("2. Ropa: 20% de descuento")
print("3. Juguetes: 15% de descuento")
print("4. Otros: Sin descuento\n")

# Definir los descuentos por categoría
descuentos = {
    "electrónica": 10,  # 10% descuento
    "ropa": 20,         # 20% descuento
    "juguetes": 15,     # 15% descuento
    "otros": 0          # Sin descuento
}

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del artículo
    articulo = input("Introduce el nombre del artículo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de artículos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del artículo
    try:
        precio_original = float(input(f"Introduce el precio original del artículo '{articulo}': €"))
    except ValueError:
        print("Error: Por favor, introduce un número válido para el precio.")
        continue

    # Paso 3: Solicitar la categoría del artículo
    categoria = input(f"Introduce la categoría del artículo '{articulo}' (electrónica, ropa, juguetes, otros): ").strip().lower()

    # Paso 4: Verificar si la categoría es válida y aplicar el descuento correspondiente
    if categoria not in descuentos:
        print("Categoría no válida. Se aplicará la categoría 'otros' sin descuento.")
        categoria = "otros"

    porcentaje_descuento = descuentos[categoria]
    descuento = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - descuento

    # Mostrar el precio final del artículo
    print(f"El precio final del artículo '{articulo}' es: €{precio_final:.2f} (Descuento aplicado: {porcentaje_descuento}%)\n")

    # Sumar al monto total
    monto_total += precio_final

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: €{monto_total:.2f}")
```

### Crea un nuevo programa que aplique descuentos automáticos en función de la fecha actual. Aplica un descuento del 10% en compras realizadas entre lunes y viernes o un descuento del 15% en compras realizadas durante los fines de semana.

```python
from datetime import datetime

# Función para determinar el descuento basado en el día de la semana
def obtener_descuento():
    # Obtener el día actual (0=lunes, 6=domingo)
    dia_actual = datetime.now().weekday()
    if dia_actual in range(0, 5):  # Lunes a Viernes
        return 10
    elif dia_actual in (5, 6):    # Sábado y Domingo
        return 15

# Bienvenida al usuario
print("Bienvenido a la tienda en línea.")
print("Descuentos automáticos según el día de la semana:")
print("- 10% de descuento: Lunes a Viernes")
print("- 15% de descuento: Sábado y Domingo\n")

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del artículo
    articulo = input("Introduce el nombre del artículo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de artículos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del artículo
    try:
        precio_original = float(input(f"Introduce el precio original del artículo '{articulo}': €"))
    except ValueError:
        print("Error: Por favor, introduce un número válido para el precio.")
        continue

    # Obtener el descuento automático según el día de la semana
    porcentaje_descuento = obtener_descuento()
    descuento = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - descuento

    # Mostrar el precio final del artículo
    print(f"El precio final del artículo '{articulo}' es: €{precio_final:.2f} (Descuento aplicado: {porcentaje_descuento}%)\n")

    # Sumar al monto total
    monto_total += precio_final

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: €{monto_total:.2f}")
```

### Crea un programa nuevo que implemente un sistema de membresía donde los usuarios registrados obtengan un descuento adicional del 5%. Solicita al usuario que ingrese su nombre de usuario y contraseña para verificar su membresía y aplique el descuento adicional si corresponde. 

📝 Nota: Para este ejercicio solo tendremos un usuario registrado. El usuario es `USUARIO` y la contraseña es  `USUARIO123`. 


```python
from datetime import datetime

# Función para determinar el descuento basado en el día de la semana
def obtener_descuento():
    dia_actual = datetime.now().weekday()
    if dia_actual in range(0, 5):  # Lunes a Viernes
        return 10
    elif dia_actual in (5, 6):    # Sábado y Domingo
        return 15

# Datos del usuario registrado
usuario_registrado = "USUARIO"
contrasena_registrada = "USUARIO123"

# Verificar membresía
print("Bienvenido a la tienda en línea.")
print("Para los miembros registrados, aplicamos un descuento adicional del 5%.\n")

es_miembro = False
usuario = input("Introduce tu nombre de usuario: ").strip()
contrasena = input("Introduce tu contraseña: ").strip()

if usuario == usuario_registrado and contrasena == contrasena_registrada:
    print("¡Autenticación exitosa! Se aplicará un descuento adicional del 5% por ser miembro.")
    es_miembro = True
else:
    print("Nombre de usuario o contraseña incorrectos. Continuarás sin el descuento adicional.\n")

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del artículo
    articulo = input("Introduce el nombre del artículo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de artículos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del artículo
    try:
        precio_original = float(input(f"Introduce el precio original del artículo '{articulo}': €"))
    except ValueError:
        print("Error: Por favor, introduce un número válido para el precio.")
        continue

    # Obtener el descuento automático según el día de la semana
    porcentaje_descuento = obtener_descuento()
    descuento = precio_original * (porcentaje_descuento / 100)

    # Aplicar descuento adicional para miembros
    if es_miembro:
        descuento_adicional = (precio_original - descuento) * 0.05
        descuento += descuento_adicional
        print(f"Se ha aplicado un descuento adicional del 5% para miembros: €{descuento_adicional:.2f}")

    precio_final = precio_original - descuento

    # Mostrar el precio final del artículo
    print(f"El precio final del artículo '{articulo}' es: €{precio_final:.2f} (Descuento total aplicado: {porcentaje_descuento + (5 if es_miembro else 0)}%)\n")

    # Sumar al monto total
    monto_total += precio_final

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: €{monto_total:.2f}")
```

### Crea un nuevo programa que ofrezca un descuento del 10% adicional si el usuario compra más de 5 artículos del mismo tipo. Debes verificar la cantidad de productos comprados y aplicar el descuento adicional si corresponde.


```python
# Programa para calcular el precio final con descuentos adicionales por cantidad de artículos

print("Bienvenido a la tienda en línea.")
print("Descuentos disponibles:")
print("- Descuento base: 10% de lunes a viernes o 15% los fines de semana.")
print("- Descuento adicional: 10% si compras más de 5 unidades del mismo artículo.\n")

from datetime import datetime

# Función para determinar el descuento basado en el día de la semana
def obtener_descuento():
    dia_actual = datetime.now().weekday()
    if dia_actual in range(0, 5):  # Lunes a Viernes
        return 10
    elif dia_actual in (5, 6):    # Sábado y Domingo
        return 15

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del artículo
    articulo = input("Introduce el nombre del artículo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de artículos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del artículo
    try:
        precio_original = float(input(f"Introduce el precio original del artículo '{articulo}': €"))
    except ValueError:
        print("Error: Por favor, introduce un número válido para el precio.")
        continue

    # Paso 3: Solicitar la cantidad de unidades compradas
    try:
        cantidad = int(input(f"¿Cuántas unidades de '{articulo}' estás comprando? "))
    except ValueError:
        print("Error: Por favor, introduce un número entero válido para la cantidad.")
        continue

    # Obtener el descuento base según el día de la semana
    porcentaje_descuento = obtener_descuento()
    descuento_base = precio_original * (porcentaje_descuento / 100)

    # Aplicar descuento adicional por cantidad
    descuento_adicional = 0
    if cantidad > 5:
        descuento_adicional = (precio_original - descuento_base) * 0.10
        print(f"Se ha aplicado un descuento adicional del 10% por comprar más de 5 unidades: €{descuento_adicional:.2f} por unidad.")

    # Calcular el precio final por unidad y luego el total para esa cantidad
    precio_unitario_final = precio_original - descuento_base - descuento_adicional
    precio_total_articulo = precio_unitario_final * cantidad

    # Mostrar el precio final del artículo
    print(f"El precio final del artículo '{articulo}' (por unidad) es: €{precio_unitario_final:.2f}")
    print(f"El total para {cantidad} unidades de '{articulo}' es: €{precio_total_articulo:.2f}\n")

    # Sumar al monto total
    monto_total += precio_total_articulo

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: €{monto_total:.2f}")
```

### Crea un nuevo programa que ofrezca descuentos progresivos en función del volumen de compra. Aplica un descuento del 5% para compras entre 100€ y 200€, un descuento del 10% para compras entre 200€ y 500€ y un descuento del 15% para compras superiores a 500€.

```python
print("Bienvenido a la tienda en línea.")
print("Descuentos progresivos por volumen de compra:")
print("- 5% de descuento: Total entre 100€ y 200€")
print("- 10% de descuento: Total entre 200€ y 500€")
print("- 15% de descuento: Total superior a 500€\n")

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del artículo
    articulo = input("Introduce el nombre del artículo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de artículos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del artículo
    try:
        precio_original = float(input(f"Introduce el precio original del artículo '{articulo}': €"))
    except ValueError:
        print("Error: Por favor, introduce un número válido para el precio.")
        continue

    # Paso 3: Solicitar la cantidad de unidades compradas
    try:
        cantidad = int(input(f"¿Cuántas unidades de '{articulo}' estás comprando? "))
    except ValueError:
        print("Error: Por favor, introduce un número entero válido para la cantidad.")
        continue

    # Calcular el precio total para este artículo
    precio_total_articulo = precio_original * cantidad

    # Mostrar el precio total del artículo
    print(f"El total para {cantidad} unidades de '{articulo}' es: €{precio_total_articulo:.2f}\n")

    # Sumar al monto total
    monto_total += precio_total_articulo

# Aplicar el descuento progresivo según el monto total
if 100 <= monto_total < 200:
    porcentaje_descuento = 5
elif 200 <= monto_total <= 500:
    porcentaje_descuento = 10
elif monto_total > 500:
    porcentaje_descuento = 15
else:
    porcentaje_descuento = 0

# Calcular el descuento y el monto final
descuento = monto_total * (porcentaje_descuento / 100)
monto_final = monto_total - descuento

# Mostrar el resultado final
print(f"\nEl monto total antes del descuento es: €{monto_total:.2f}")
if porcentaje_descuento > 0:
    print(f"Se ha aplicado un descuento del {porcentaje_descuento}%: €{descuento:.2f}")
print(f"El monto final después del descuento es: €{monto_final:.2f}")
```

### Crea un nuevo programa  que ofrezca un descuento adicional del 3% si el usuario paga con tarjeta de crédito o débito. Solicita al usuario que ingrese el método de pago y el descuento original, por último aplica el descuento si corresponde.

```python
print("Bienvenido a la tienda en línea.")
print("Obtén un descuento adicional del 3% si pagas con tarjeta de crédito o débito.\n")

# Paso 1: Solicitar el monto total antes del descuento
try:
    monto_original = float(input("Introduce el monto total antes de aplicar descuentos: €"))
except ValueError:
    print("Error: Por favor, introduce un número válido para el monto total.")
    exit()

# Paso 2: Solicitar el descuento original aplicado
try:
    descuento_original = float(input("Introduce el descuento original en porcentaje (por ejemplo, 10 para 10%): "))
except ValueError:
    print("Error: Por favor, introduce un número válido para el descuento.")
    exit()

# Calcular el monto después del descuento original
descuento_aplicado = monto_original * (descuento_original / 100)
monto_con_descuento = monto_original - descuento_aplicado

# Mostrar el monto con el descuento original
print(f"\nEl monto después del descuento original ({descuento_original}%): €{monto_con_descuento:.2f}")

# Paso 3: Solicitar el método de pago
metodo_pago = input("¿Cuál es tu método de pago? (escribe 'tarjeta de crédito', 'tarjeta de débito' o 'efectivo'): ").strip().lower()

# Verificar si el usuario paga con tarjeta y aplicar descuento adicional
descuento_adicional = 0
if metodo_pago in ["tarjeta de crédito", "tarjeta de débito"]:
    descuento_adicional = monto_con_descuento * 0.03
    print(f"Se ha aplicado un descuento adicional del 3% por pagar con {metodo_pago}.")

# Calcular el monto final después de todos los descuentos
monto_final = monto_con_descuento - descuento_adicional

# Mostrar el monto final
print(f"El monto final después de todos los descuentos es: €{monto_final:.2f}")
```
