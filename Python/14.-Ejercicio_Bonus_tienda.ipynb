En este ejercicio, se te pedir√° que escribas un programa en Python que utilice condicionales para determinar el monto final de una compra en una tienda en l√≠nea, despu√©s de aplicar un descuento. El programa debe hacer lo siguiente:

1. Solicita al usuario que ingrese el precio original de un art√≠culo.

2. Pregunta al usuario si tiene un cup√≥n de descuento (respuesta s√≠ o no).

3. Si el usuario responde que s√≠, solicita que ingrese el valor del cup√≥n de descuento.

4. Aplica el descuento al precio original del art√≠culo, siempre y cuando el valor del cup√≥n sea v√°lido (es decir, mayor a cero). Por ejemplo, descuento de 15‚Ç¨. 

5. Muestra el precio final de la compra, teniendo en cuenta el descuento aplicado o sin √©l. 

6. Recuerda utilizar estructuras de control de flujo como if, elif y else para llevar a cabo estas acciones en tu programa de Python.

```python
print("Bienvenido a la tienda en l√≠nea.")

# Paso 1: Solicitar el precio original del art√≠culo
precio_original = float(input("Por favor, introduce el precio original del art√≠culo: ‚Ç¨"))

# Paso 2: Preguntar si el usuario tiene un cup√≥n de descuento
tiene_cupon = input("¬øTienes un cup√≥n de descuento? (responde 'si' o 'no'): ").strip().lower()

# Paso 3 y 4: Verificar si hay cup√≥n y calcular el descuento
if tiene_cupon == "si":
    valor_cupon = float(input("Introduce el valor del cup√≥n de descuento (‚Ç¨): "))
    
    if valor_cupon > 0:
        # Calcular el precio final con el descuento aplicado
        precio_final = precio_original - valor_cupon
        
        # Asegurarnos de que el precio final no sea negativo
        if precio_final < 0:
            precio_final = 0  # El precio m√≠nimo es 0
    else:
        print("El valor del cup√≥n no es v√°lido. Se aplicar√° el precio original.")
        precio_final = precio_original
else:
    # No hay cup√≥n de descuento
    precio_final = precio_original

# Paso 5: Mostrar el precio final de la compra
print(f"El precio final de la compra es: {precio_final:.2f}‚Ç¨")
```

### Modifica el programa anterior agregando un l√≠mite al descuento, donde el descuento m√≠nimo sea del 5% y el m√°ximo del 50%. Si el usuario ingresa un porcentaje fuera de este rango, notif√≠quele y aplique el descuento m√≠nimo o m√°ximo seg√∫n sea el caso.

```python
print("Bienvenido a la tienda en l√≠nea.")

# Paso 1: Solicitar el precio original del art√≠culo
precio_original = float(input("Por favor, introduce el precio original del art√≠culo: ‚Ç¨"))

# Paso 2: Preguntar si el usuario tiene un porcentaje de descuento
tiene_descuento = input("¬øTienes un porcentaje de descuento? (responde 'si' o 'no'): ").strip().lower()

# L√≠mite de los descuentos
descuento_minimo = 5
descuento_maximo = 50

# Paso 3 y 4: Verificar si hay descuento y calcular el precio final
if tiene_descuento == "si":
    porcentaje_descuento = float(input("Introduce el porcentaje de descuento (por ejemplo, 30 para 30%): "))
    
    # Ajustar el descuento si est√° fuera del rango permitido
    if porcentaje_descuento < descuento_minimo:
        print(f"El porcentaje ingresado es menor al m√≠nimo permitido ({descuento_minimo}%). Se aplicar√° el descuento m√≠nimo.")
        porcentaje_descuento = descuento_minimo
    elif porcentaje_descuento > descuento_maximo:
        print(f"El porcentaje ingresado supera el m√°ximo permitido ({descuento_maximo}%). Se aplicar√° el descuento m√°ximo.")
        porcentaje_descuento = descuento_maximo

    # Calcular el precio final con el porcentaje ajustado
    descuento = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - descuento
else:
    # No hay descuento aplicado
    precio_final = precio_original

# Paso 5: Mostrar el precio final de la compra
print(f"El precio final de la compra es: ‚Ç¨{precio_final:.2f}")
```

### Crea un programa nuevo que permita al usuario ingresar varios art√≠culos, cada uno con su precio original y cup√≥n de descuento. Aplica los descuentos correspondientes a cada producto y muestra el monto total final de la compra.

```python
print("Bienvenido a la tienda en l√≠nea.")
print("Podr√°s ingresar varios art√≠culos con sus precios y descuentos.")
print("Cuando termines de ingresar los art√≠culos, escribe 'fin'.")

# Inicializar variables
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del art√≠culo
    articulo = input("Introduce el nombre del art√≠culo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de art√≠culos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del art√≠culo
    try:
        precio_original = float(input(f"Introduce el precio original del art√≠culo '{articulo}': ‚Ç¨"))
    except ValueError:
        print("Error: Por favor, introduce un n√∫mero v√°lido para el precio.")
        continue

    # Paso 3: Preguntar si el art√≠culo tiene un porcentaje de descuento
    tiene_descuento = input(f"¬øEl art√≠culo '{articulo}' tiene un descuento? (responde 'si' o 'no'): ").strip().lower()

    # Inicializar precio final para este art√≠culo
    precio_final = precio_original

    if tiene_descuento == "si":
        try:
            # Paso 4: Solicitar el porcentaje de descuento
            porcentaje_descuento = float(input("Introduce el porcentaje de descuento (por ejemplo, 30 para 30%): "))

            # Aplicar validaci√≥n del porcentaje de descuento (m√≠nimo 5%, m√°ximo 50%)
            if porcentaje_descuento < 5:
                print("El descuento es menor al m√≠nimo permitido (5%). Se aplicar√° el descuento m√≠nimo del 5%.")
                porcentaje_descuento = 5
            elif porcentaje_descuento > 50:
                print("El descuento supera el m√°ximo permitido (50%). Se aplicar√° el descuento m√°ximo del 50%.")
                porcentaje_descuento = 50

            # Calcular el descuento y el precio final
            descuento = precio_original * (porcentaje_descuento / 100)
            precio_final -= descuento

        except ValueError:
            print("Error: Por favor, introduce un n√∫mero v√°lido para el porcentaje de descuento.")
            continue

    # Agregar el precio final del art√≠culo al monto total
    monto_total += precio_final
    print(f"\nEl precio final del art√≠culo '{articulo}' es: ‚Ç¨{precio_final:.2f}\n")

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: ‚Ç¨{monto_total:.2f}")
```

### Crea un nuevo programa que permita aplicar descuentos por categor√≠as. En este caso, nuestra tienda aplicar√° descuentos en funci√≥n de la categor√≠a (por ejemplo, electr√≥nica, ropa, juguetes). Pregunta al usuario a que categor√≠a pertenece su producto y aplica el descuento correspondiente a la categor√≠a. Debes solicitar al usuario que ingrese la categor√≠a del articulo y el precio para aplicarle el descuento. 

```python
print("Bienvenido a la tienda en l√≠nea.")
print("Nuestra tienda ofrece descuentos seg√∫n las categor√≠as:")
print("1. Electr√≥nica: 10% de descuento")
print("2. Ropa: 20% de descuento")
print("3. Juguetes: 15% de descuento")
print("4. Otros: Sin descuento\n")

# Definir los descuentos por categor√≠a
descuentos = {
    "electr√≥nica": 10,  # 10% descuento
    "ropa": 20,         # 20% descuento
    "juguetes": 15,     # 15% descuento
    "otros": 0          # Sin descuento
}

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del art√≠culo
    articulo = input("Introduce el nombre del art√≠culo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de art√≠culos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del art√≠culo
    try:
        precio_original = float(input(f"Introduce el precio original del art√≠culo '{articulo}': ‚Ç¨"))
    except ValueError:
        print("Error: Por favor, introduce un n√∫mero v√°lido para el precio.")
        continue

    # Paso 3: Solicitar la categor√≠a del art√≠culo
    categoria = input(f"Introduce la categor√≠a del art√≠culo '{articulo}' (electr√≥nica, ropa, juguetes, otros): ").strip().lower()

    # Paso 4: Verificar si la categor√≠a es v√°lida y aplicar el descuento correspondiente
    if categoria not in descuentos:
        print("Categor√≠a no v√°lida. Se aplicar√° la categor√≠a 'otros' sin descuento.")
        categoria = "otros"

    porcentaje_descuento = descuentos[categoria]
    descuento = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - descuento

    # Mostrar el precio final del art√≠culo
    print(f"El precio final del art√≠culo '{articulo}' es: ‚Ç¨{precio_final:.2f} (Descuento aplicado: {porcentaje_descuento}%)\n")

    # Sumar al monto total
    monto_total += precio_final

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: ‚Ç¨{monto_total:.2f}")
```

### Crea un nuevo programa que aplique descuentos autom√°ticos en funci√≥n de la fecha actual. Aplica un descuento del 10% en compras realizadas entre lunes y viernes o un descuento del 15% en compras realizadas durante los fines de semana.

```python
from datetime import datetime

# Funci√≥n para determinar el descuento basado en el d√≠a de la semana
def obtener_descuento():
    # Obtener el d√≠a actual (0=lunes, 6=domingo)
    dia_actual = datetime.now().weekday()
    if dia_actual in range(0, 5):  # Lunes a Viernes
        return 10
    elif dia_actual in (5, 6):    # S√°bado y Domingo
        return 15

# Bienvenida al usuario
print("Bienvenido a la tienda en l√≠nea.")
print("Descuentos autom√°ticos seg√∫n el d√≠a de la semana:")
print("- 10% de descuento: Lunes a Viernes")
print("- 15% de descuento: S√°bado y Domingo\n")

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del art√≠culo
    articulo = input("Introduce el nombre del art√≠culo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de art√≠culos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del art√≠culo
    try:
        precio_original = float(input(f"Introduce el precio original del art√≠culo '{articulo}': ‚Ç¨"))
    except ValueError:
        print("Error: Por favor, introduce un n√∫mero v√°lido para el precio.")
        continue

    # Obtener el descuento autom√°tico seg√∫n el d√≠a de la semana
    porcentaje_descuento = obtener_descuento()
    descuento = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - descuento

    # Mostrar el precio final del art√≠culo
    print(f"El precio final del art√≠culo '{articulo}' es: ‚Ç¨{precio_final:.2f} (Descuento aplicado: {porcentaje_descuento}%)\n")

    # Sumar al monto total
    monto_total += precio_final

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: ‚Ç¨{monto_total:.2f}")
```

### Crea un programa nuevo que implemente un sistema de membres√≠a donde los usuarios registrados obtengan un descuento adicional del 5%. Solicita al usuario que ingrese su nombre de usuario y contrase√±a para verificar su membres√≠a y aplique el descuento adicional si corresponde. 

üìù Nota: Para este ejercicio solo tendremos un usuario registrado. El usuario es `USUARIO` y la contrase√±a es  `USUARIO123`. 


```python
from datetime import datetime

# Funci√≥n para determinar el descuento basado en el d√≠a de la semana
def obtener_descuento():
    dia_actual = datetime.now().weekday()
    if dia_actual in range(0, 5):  # Lunes a Viernes
        return 10
    elif dia_actual in (5, 6):    # S√°bado y Domingo
        return 15

# Datos del usuario registrado
usuario_registrado = "USUARIO"
contrasena_registrada = "USUARIO123"

# Verificar membres√≠a
print("Bienvenido a la tienda en l√≠nea.")
print("Para los miembros registrados, aplicamos un descuento adicional del 5%.\n")

es_miembro = False
usuario = input("Introduce tu nombre de usuario: ").strip()
contrasena = input("Introduce tu contrase√±a: ").strip()

if usuario == usuario_registrado and contrasena == contrasena_registrada:
    print("¬°Autenticaci√≥n exitosa! Se aplicar√° un descuento adicional del 5% por ser miembro.")
    es_miembro = True
else:
    print("Nombre de usuario o contrase√±a incorrectos. Continuar√°s sin el descuento adicional.\n")

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del art√≠culo
    articulo = input("Introduce el nombre del art√≠culo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de art√≠culos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del art√≠culo
    try:
        precio_original = float(input(f"Introduce el precio original del art√≠culo '{articulo}': ‚Ç¨"))
    except ValueError:
        print("Error: Por favor, introduce un n√∫mero v√°lido para el precio.")
        continue

    # Obtener el descuento autom√°tico seg√∫n el d√≠a de la semana
    porcentaje_descuento = obtener_descuento()
    descuento = precio_original * (porcentaje_descuento / 100)

    # Aplicar descuento adicional para miembros
    if es_miembro:
        descuento_adicional = (precio_original - descuento) * 0.05
        descuento += descuento_adicional
        print(f"Se ha aplicado un descuento adicional del 5% para miembros: ‚Ç¨{descuento_adicional:.2f}")

    precio_final = precio_original - descuento

    # Mostrar el precio final del art√≠culo
    print(f"El precio final del art√≠culo '{articulo}' es: ‚Ç¨{precio_final:.2f} (Descuento total aplicado: {porcentaje_descuento + (5 if es_miembro else 0)}%)\n")

    # Sumar al monto total
    monto_total += precio_final

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: ‚Ç¨{monto_total:.2f}")
```

### Crea un nuevo programa que ofrezca un descuento del 10% adicional si el usuario compra m√°s de 5 art√≠culos del mismo tipo. Debes verificar la cantidad de productos comprados y aplicar el descuento adicional si corresponde.


```python
# Programa para calcular el precio final con descuentos adicionales por cantidad de art√≠culos

print("Bienvenido a la tienda en l√≠nea.")
print("Descuentos disponibles:")
print("- Descuento base: 10% de lunes a viernes o 15% los fines de semana.")
print("- Descuento adicional: 10% si compras m√°s de 5 unidades del mismo art√≠culo.\n")

from datetime import datetime

# Funci√≥n para determinar el descuento basado en el d√≠a de la semana
def obtener_descuento():
    dia_actual = datetime.now().weekday()
    if dia_actual in range(0, 5):  # Lunes a Viernes
        return 10
    elif dia_actual in (5, 6):    # S√°bado y Domingo
        return 15

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del art√≠culo
    articulo = input("Introduce el nombre del art√≠culo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de art√≠culos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del art√≠culo
    try:
        precio_original = float(input(f"Introduce el precio original del art√≠culo '{articulo}': ‚Ç¨"))
    except ValueError:
        print("Error: Por favor, introduce un n√∫mero v√°lido para el precio.")
        continue

    # Paso 3: Solicitar la cantidad de unidades compradas
    try:
        cantidad = int(input(f"¬øCu√°ntas unidades de '{articulo}' est√°s comprando? "))
    except ValueError:
        print("Error: Por favor, introduce un n√∫mero entero v√°lido para la cantidad.")
        continue

    # Obtener el descuento base seg√∫n el d√≠a de la semana
    porcentaje_descuento = obtener_descuento()
    descuento_base = precio_original * (porcentaje_descuento / 100)

    # Aplicar descuento adicional por cantidad
    descuento_adicional = 0
    if cantidad > 5:
        descuento_adicional = (precio_original - descuento_base) * 0.10
        print(f"Se ha aplicado un descuento adicional del 10% por comprar m√°s de 5 unidades: ‚Ç¨{descuento_adicional:.2f} por unidad.")

    # Calcular el precio final por unidad y luego el total para esa cantidad
    precio_unitario_final = precio_original - descuento_base - descuento_adicional
    precio_total_articulo = precio_unitario_final * cantidad

    # Mostrar el precio final del art√≠culo
    print(f"El precio final del art√≠culo '{articulo}' (por unidad) es: ‚Ç¨{precio_unitario_final:.2f}")
    print(f"El total para {cantidad} unidades de '{articulo}' es: ‚Ç¨{precio_total_articulo:.2f}\n")

    # Sumar al monto total
    monto_total += precio_total_articulo

# Paso 5: Mostrar el monto total de la compra
print(f"\nEl monto total final de la compra es: ‚Ç¨{monto_total:.2f}")
```

### Crea un nuevo programa que ofrezca descuentos progresivos en funci√≥n del volumen de compra. Aplica un descuento del 5% para compras entre 100‚Ç¨ y 200‚Ç¨, un descuento del 10% para compras entre 200‚Ç¨ y 500‚Ç¨ y un descuento del 15% para compras superiores a 500‚Ç¨.

```python
print("Bienvenido a la tienda en l√≠nea.")
print("Descuentos progresivos por volumen de compra:")
print("- 5% de descuento: Total entre 100‚Ç¨ y 200‚Ç¨")
print("- 10% de descuento: Total entre 200‚Ç¨ y 500‚Ç¨")
print("- 15% de descuento: Total superior a 500‚Ç¨\n")

# Inicializar monto total
monto_total = 0

while True:
    # Paso 1: Solicitar el nombre del art√≠culo
    articulo = input("Introduce el nombre del art√≠culo (o escribe 'fin' para terminar): ").strip()
    
    # Finalizar el ingreso de art√≠culos
    if articulo.lower() == "fin":
        break

    # Paso 2: Solicitar el precio original del art√≠culo
    try:
        precio_original = float(input(f"Introduce el precio original del art√≠culo '{articulo}': ‚Ç¨"))
    except ValueError:
        print("Error: Por favor, introduce un n√∫mero v√°lido para el precio.")
        continue

    # Paso 3: Solicitar la cantidad de unidades compradas
    try:
        cantidad = int(input(f"¬øCu√°ntas unidades de '{articulo}' est√°s comprando? "))
    except ValueError:
        print("Error: Por favor, introduce un n√∫mero entero v√°lido para la cantidad.")
        continue

    # Calcular el precio total para este art√≠culo
    precio_total_articulo = precio_original * cantidad

    # Mostrar el precio total del art√≠culo
    print(f"El total para {cantidad} unidades de '{articulo}' es: ‚Ç¨{precio_total_articulo:.2f}\n")

    # Sumar al monto total
    monto_total += precio_total_articulo

# Aplicar el descuento progresivo seg√∫n el monto total
if 100 <= monto_total < 200:
    porcentaje_descuento = 5
elif 200 <= monto_total <= 500:
    porcentaje_descuento = 10
elif monto_total > 500:
    porcentaje_descuento = 15
else:
    porcentaje_descuento = 0

# Calcular el descuento y el monto final
descuento = monto_total * (porcentaje_descuento / 100)
monto_final = monto_total - descuento

# Mostrar el resultado final
print(f"\nEl monto total antes del descuento es: ‚Ç¨{monto_total:.2f}")
if porcentaje_descuento > 0:
    print(f"Se ha aplicado un descuento del {porcentaje_descuento}%: ‚Ç¨{descuento:.2f}")
print(f"El monto final despu√©s del descuento es: ‚Ç¨{monto_final:.2f}")
```

### Crea un nuevo programa  que ofrezca un descuento adicional del 3% si el usuario paga con tarjeta de cr√©dito o d√©bito. Solicita al usuario que ingrese el m√©todo de pago y el descuento original, por √∫ltimo aplica el descuento si corresponde.

```python
print("Bienvenido a la tienda en l√≠nea.")
print("Obt√©n un descuento adicional del 3% si pagas con tarjeta de cr√©dito o d√©bito.\n")

# Paso 1: Solicitar el monto total antes del descuento
try:
    monto_original = float(input("Introduce el monto total antes de aplicar descuentos: ‚Ç¨"))
except ValueError:
    print("Error: Por favor, introduce un n√∫mero v√°lido para el monto total.")
    exit()

# Paso 2: Solicitar el descuento original aplicado
try:
    descuento_original = float(input("Introduce el descuento original en porcentaje (por ejemplo, 10 para 10%): "))
except ValueError:
    print("Error: Por favor, introduce un n√∫mero v√°lido para el descuento.")
    exit()

# Calcular el monto despu√©s del descuento original
descuento_aplicado = monto_original * (descuento_original / 100)
monto_con_descuento = monto_original - descuento_aplicado

# Mostrar el monto con el descuento original
print(f"\nEl monto despu√©s del descuento original ({descuento_original}%): ‚Ç¨{monto_con_descuento:.2f}")

# Paso 3: Solicitar el m√©todo de pago
metodo_pago = input("¬øCu√°l es tu m√©todo de pago? (escribe 'tarjeta de cr√©dito', 'tarjeta de d√©bito' o 'efectivo'): ").strip().lower()

# Verificar si el usuario paga con tarjeta y aplicar descuento adicional
descuento_adicional = 0
if metodo_pago in ["tarjeta de cr√©dito", "tarjeta de d√©bito"]:
    descuento_adicional = monto_con_descuento * 0.03
    print(f"Se ha aplicado un descuento adicional del 3% por pagar con {metodo_pago}.")

# Calcular el monto final despu√©s de todos los descuentos
monto_final = monto_con_descuento - descuento_adicional

# Mostrar el monto final
print(f"El monto final despu√©s de todos los descuentos es: ‚Ç¨{monto_final:.2f}")
```
